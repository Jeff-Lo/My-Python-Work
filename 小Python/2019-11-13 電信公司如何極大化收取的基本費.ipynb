# 找出消費者剩餘
# 根據電信公司在市場上對於手機通話的均衡收費是每小時多少元極大化基本費
### 消費者剩餘便是可收基本費的最大值

```python
market_price_per_hour=input("請輸入市場均衡收費是每小時多少元:")
market_price_per_hour=int(market_price_per_hour)

def Q(y):
    P=y
    Q=(40-P)/0.2
    return Q

if(0<=market_price_per_hour<=40):
    x1=[0,200]
    y1=[40,0]
    plt.plot(x1,y1)
    x2=[0,Q(market_price_per_hour)]
    y2=[market_price_per_hour,market_price_per_hour]
    plt.plot(x2,y2)

else:
    print("A impossible transaction.")

con_sur=(40-market_price_per_hour)*Q(market_price_per_hour)/2
print("消費者剩餘是",con_sur)

fundamental=con_sur
print("電信公司對於手機通話的基本費最多可在",fundamental,"以下")
```
