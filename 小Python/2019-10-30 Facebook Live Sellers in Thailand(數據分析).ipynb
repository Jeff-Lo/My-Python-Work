# 讀入資料(Facebook Live Sellers in Thailand)

```python
df=pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/00488/Live.csv")
```

```python
df.head()
```

# 刪掉欄位

```python
df.drop(["status_id","Column1","Column2","Column3","Column4"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.index=range(1,len(df)+1)
```
