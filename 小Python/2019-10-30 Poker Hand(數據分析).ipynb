```python
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/poker/poker-hand-testing.data")
```

```python
df.head()
```

# 整理資料

```python
df.columns=list(["第1張牌的花色","第1張牌的數字","第2張牌的花色","第2張牌的數字","第3張牌的花色","第3張牌的數字","第4張牌的花色","第4張牌的數字","第5張牌的花色","第5張牌的數字","類型"])
```

```python
df.index=range(1,len(df)+1)
```

```python
df.head()
```

```python
df.tail()
```

# 依照類型來看

```python
df2=df.sort_values(by="類型",ascending=False)
```

# 查詢牌型情報

```python
overall=len(df2)
nothing_in_hand=len(df2[df2.類型==0].index)/overall*100
one_pair=len(df2[df2.類型==1].index)/overall*100
two_pairs=len(df2[df2.類型==2].index)/overall*100
three_of_a_kind=len(df2[df2.類型==3].index)/overall*100
straight=len(df2[df2.類型==4].index)/overall*100
flush=len(df2[df2.類型==5].index)/overall*100
full_house=len(df2[df2.類型==6].index)/overall*100
four_of_a_kind=len(df2[df2.類型==7].index)/overall*100
straight_flush=len(df2[df2.類型==8].index)/overall*100
royal_flush=len(df2[df2.類型==9].index)/overall*100
```

```python
information=input("請輸入你想得知某種牌型的情報→")
if information=="亂牌":
    print(f"拿到的機率佔{nothing_in_hand}%")
elif information=="一對":
    print(f"拿到的機率佔{one_pair}%")
elif information=="兩對":
    print(f"拿到的機率佔{two_pairs}%")
elif information=="三條":
    print(f"拿到的機率佔{three_of_a_kind}%")
elif information=="順子":
    print(f"拿到的機率佔{straight}%")
elif information=="同花":
    print(f"拿到的機率佔{flush}%")
elif information=="葫蘆":
    print(f"拿到的機率佔{full_house}%")
elif information=="鐵支":
    print(f"拿到的機率佔{four_of_a_kind}%")
elif information=="同花順":
    print(f"拿到的機率佔{straight_flush}%")
elif information=="皇家同花順":
    print(f"拿到的機率佔{royal_flush}%")
else:
    print("很抱歉，並無此情報。")
```
