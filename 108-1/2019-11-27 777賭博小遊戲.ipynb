```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#1枚代幣 = 100元

#玩一次投2枚

def pull_down(position = 3 , pattern = None):
    
    print("----------拉下把手----------")
    
    if pattern == None:
        
        pattern = []
    
    for i in range(position):
        
        point = np.random.choice(["♥","✿",7,"✧","♛"])
        
        pattern.append(point)
    
    return pattern

def start_game():
    
    your_money = 1000
        
    print("----------遊戲開始----------")
        
    print(f"你的初始本金是{your_money}元。")
        
    token_in = input("請輸入「in」投入2枚代幣:")
        
    if (token_in == "in"):
        
        your_money = your_money - 100 * 2
        
        patterns = pull_down()
        
        youWin = patterns == [7,7,7]
        
        if youWin:
                
            your_money = your_money + 1000000
                
            print(patterns)
            
            print(f"恭喜幸運的你可以拿到一百萬元!!!你現在共有{your_money}元。\n")
                
            play_or_leave = "----------系統錯誤，請重新輸入。----------"
            
            while (play_or_leave == "----------系統錯誤，請重新輸入。----------"):
                
                play_or_leave = input("\n你還要在玩一次嗎? Y/N:\n")
            
                if (play_or_leave == "Y"):
                    
                    print("\n")
                    
                    start_game()
                    
                elif (play_or_leave == "N"):
                    
                    print("謝謝您~歡迎下次再來玩!")
                    
                else:
                    
                    print("----------系統錯誤，請重新輸入。----------")
        
        else:
                
            print(patterns)
                
            print(f"真可惜，下次請繼續努力~你現在剩下{your_money}元。\n")
            
            play_or_leave = "----------系統錯誤，請重新輸入。----------"
            
            while (play_or_leave == "----------系統錯誤，請重新輸入。----------"):
                
                play_or_leave = input("\n你還要在玩一次嗎? Y/N:")
                
                if (play_or_leave == "Y"):
                    
                    print("\n")
                    
                    start_game()
                    
                elif (play_or_leave == "N"):
                    
                    print("謝謝您~歡迎下次再來玩!")
                    
                else:
                    
                    play_or_leave = "----------系統錯誤，請重新輸入。----------"
                    
                    print("----------系統錯誤，請重新輸入。----------")
            
    else:
            
        print("----------系統錯誤，請重新輸入。----------\n")
            
        start_game()
```
