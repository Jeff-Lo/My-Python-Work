```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def roll_dice(dice = 3 , points = None):
    
    print("----------搖骰子----------")
    
    if (points == None):
        
        points = []
    
    for i in range(dice):
        
        number = np.random.randint(1,7)
        
        points.append(number)
    
    return points

def roll_final(total):
    
    big = 11 <= total <= 18
    
    small = 3 <= total <= 10
    
    if big:
        
        return "大"
    
    elif small:
        
        return "小"

def start_game():
    
    your_money = 1000
    
    while your_money > 0:
        
        print("----------遊戲開始----------")
        
        print(f"你的初始本金是{your_money}元。")
        
        choice = ["大","小"]
        
        your_choice = input("請輸入押「大」還是「小」:")
        
        if your_choice in choice:
            
            your_bet = input("請輸入「下注金額」:")
            
            your_bet = int(your_bet)
            
            your_odds = input("請設定賠率數字:")
            
            your_odds = float(your_odds)
            
            points = roll_dice()
            
            total = sum(points)
            
            youWin = your_choice == roll_final(total)
            
            if youWin:
                
                your_money = your_money + your_bet * your_odds
                
                print("骰子點數:",points)
                
                print(f"恭喜你贏了{your_bet * your_odds}元，你現在有{your_money}的本金。\n")
            
            else:
                
                your_money = your_money-your_bet * your_odds
                
                print("骰子點數:",points)
                
                print(f"真可惜你輸了{your_bet * your_odds}元，你現在剩下{your_money}的本金。\n")
        
        else:
            
            print("----------系統錯誤，請重新輸入。----------\n")
    
    else:
        
        print("----------破產啦!!!遊戲結束!!!!----------")
```

```python
start_game()
```
